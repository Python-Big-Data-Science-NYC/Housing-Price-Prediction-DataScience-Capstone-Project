```python
#coest2016alldata = pd.read_csv()

#coest2016alldata.filter(regex='2010').columns

lofc = []
for colname in coest2016alldata.filter(regex='2016').columns:
    lofc.append( colname) #.replace('2010', ''))
    #print(colname)

lofc = list(set(lofc))
lofc
```

```python
#output_names = [name for name if (name[0] =='R')]
# https://stackoverflow.com/questions/10134372/get-a-list-of-names-which-start-with-certain-letters

names = lofc
first_letters = ['R']
output_names = [name for name in names if (name[0] in first_letters)]


try:
    output_names.remove( 'RESIDUAL2016')
except ValueError:
    pass

output_names
```

```python

xpop = 'RESIDUAL2016'
    
    
    
def makedf (xpop):
    #print (xpop[:-4])
    df  = coest2016alldata.filter( regex = xpop[:-4] ) #.max()
    df = df.reset_index()
    
    df = pd.melt(df, id_vars='CTYNAME',  var_name=xpop+'Year', value_name=xpop)
    df['Year'] = df[xpop+'Year'].str[-4:]
    df['Year'] = pd.to_datetime(df['Year'], format='%Y-%m')
    del df[xpop+'Year']
    df.columns.values[0] = 'county'
    return df

RNETMIG = makedf('RNETMIG2016')
RNETMIG

dflist = []
for colnames in output_names:
    dflist.append(makedf(colnames))
    
dflist[1]
```

```python

RDEATH  = coest2016alldata.filter( regex = 'RDEATH' ) #.max()
RDEATH = RDEATH.reset_index()
RDEATH = pd.melt(RDEATH, id_vars='CTYNAME',  var_name='DeathYear', value_name='deaths')
RDEATH['Year'] = RDEATH['DeathYear'].str[6:]
RDEATH['Year'] = pd.to_datetime(RDEATH['Year'], format='%Y-%m')
del RDEATH['DeathYear']
RDEATH
```

```python
#http://rstudio-pubs-static.s3.amazonaws.com/217450_4fabdb443de345efbc493cf0161ac51d.html

DEATHS = coest2016alldata.filter( like = 'DEATHS' ) #.max()
DEATHS = DEATHS.reset_index()
DEATHS = pd.melt(DEATHS, id_vars='CTYNAME',  var_name='DeathYear', value_name='deaths')
DEATHS['Year'] = DEATHS['DeathYear'].str[6:]
DEATHS = DEATHS [DEATHS['Year'] !='2010']
DEATHS['Year'] = pd.to_datetime(DEATHS['Year'], format='%Y-%m')
del DEATHS['DeathYear']
DEATHS = DEATHS.pivot_table(index=['Year'], columns=['CTYNAME'], values='deaths').fillna(0)
DEATHS = DEATHS.pct_change()
DEATHS = DEATHS.reset_index()
DEATHS = pd.melt(DEATHS, id_vars='Year',  var_name='CTYNAME', value_name='deaths')
DEATHS
```
