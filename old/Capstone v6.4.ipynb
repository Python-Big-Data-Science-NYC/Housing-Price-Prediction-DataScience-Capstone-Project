```python
Zip_Zhvi_AllHomes_county = pd.read_csv('County_MedianListingPricePerSqft_AllHomes.csv')
Zip_Zhvi_AllHomes_county.dropna(thresh=40,inplace= True)  
```

```python
Zip_Zhvi_AllHomes_county["RegionName"] = Zip_Zhvi_AllHomes_county["RegionName"] + ' '+ Zip_Zhvi_AllHomes_county["State"]
Zip_Zhvi_AllHomes_county.head()
```

```python
Zip_Zhvi_AllHomes_countyT= Zip_Zhvi_AllHomes_county.transpose()

Zip_Zhvi_AllHomes_countyT.drop(Zip_Zhvi_AllHomes_countyT.index[1:6], inplace= True)
Zip_Zhvi_AllHomes_countyT.columns = Zip_Zhvi_AllHomes_countyT.iloc[0]
Zip_Zhvi_AllHomes_countyT.drop(Zip_Zhvi_AllHomes_countyT.index[0], inplace= True)


#Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.reset_index()
Zip_Zhvi_AllHomes_countyT.index =  pd.to_datetime(Zip_Zhvi_AllHomes_countyT.index, format='%Y-%m')
Zip_Zhvi_AllHomes_countyT['month'] = pd.DatetimeIndex(Zip_Zhvi_AllHomes_countyT.index).month

Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.loc[Zip_Zhvi_AllHomes_countyT['month'] == 1]
Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.pct_change()

del Zip_Zhvi_AllHomes_countyT['month']
# #>>> pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])
Zip_Zhvi_AllHomes_countyT
```

```python
Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.reset_index()

Zip_Zhvi_AllHomes_countyT = pd.melt(Zip_Zhvi_AllHomes_countyT,  id_vars='index', var_name='county', value_name='price')
Zip_Zhvi_AllHomes_countyT
```
