```python
import pandas as pd
```

```python

Zip_Zhvi_AllHomes = pd.read_csv('http://files.zillowstatic.com/research/public/Zip/Zip_Zhvi_AllHomes.csv')
#Zip_Zhvi_AllHomes.iloc[1,:]

```

```python
Zip_Zhvi_AllHomesT= Zip_Zhvi_AllHomes.transpose()
 #.iloc[0,:]
#df.drop()
Zip_Zhvi_AllHomesT.drop(Zip_Zhvi_AllHomesT.index[2:7], inplace= True)
Zip_Zhvi_AllHomesT.drop(Zip_Zhvi_AllHomesT.index[0], inplace= True)
Zip_Zhvi_AllHomesT.columns = Zip_Zhvi_AllHomesT.iloc[0]
Zip_Zhvi_AllHomesT.drop(Zip_Zhvi_AllHomesT.index[0], inplace= True)
Zip_Zhvi_AllHomesT = Zip_Zhvi_AllHomesT.pct_change()

Zip_Zhvi_AllHomesT = Zip_Zhvi_AllHomesT.reset_index()

#>>> pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])
Zip_Zhvi_AllHomesT
```

```python
#Zip_Zhvi_AllHomes.count()
```

```python
Zip_Zhvi_AllHomesT = pd.melt(Zip_Zhvi_AllHomesT,  id_vars='index', var_name='zip', value_name='price')
Zip_Zhvi_AllHomesT
```

```python
from datetime import datetime
datestring = "2008-12"
dt = datetime.strptime(datestring, '%Y-%m')
print (dt.year, dt.month, dt.day)
```
