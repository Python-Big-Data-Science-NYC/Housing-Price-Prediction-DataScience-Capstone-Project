```python

xpop = 'RESIDUAL2016'
    
    
    
def makedf (xpop):
    #print (xpop[:-4])
    df  = coest2016alldata.filter( regex = xpop[:-4] ) #.max()
    df = df.reset_index()
    
    df = pd.melt(df, id_vars='CTYNAME',  var_name=xpop+'Year', value_name=xpop)
    df['Year'] = df[xpop+'Year'].str[-4:]
    df['Year'] = pd.to_datetime(df['Year'], format='%Y-%m')
    del df[xpop+'Year']
    df.columns.values[0] = 'county'
    return df

RNETMIG = makedf('RNETMIG2016')
RNETMIG

dflist = []
for colnames in output_names:
    dflist.append(makedf(colnames))
    
#dflist[1]
```

```python
#Unemployment rate
df = pd.read_csv('output.csv')
df['County'] = df['County'].str.replace(' County','')
df['County'] = df['County'].str.replace('County','')
df = df.replace({"State": states})
df['County'] = df['County']+' '+ df['State']
df['Year'] = df['Month'] + ' ' +df['Year'].astype(str)
df['Year']=  pd.to_datetime(df['Year'], format='%B %Y')
del df['State']
del df['Month']
df = df.pivot_table(index=['Year'], columns=['County'], values='Rate').fillna(0)
#pivot will put county into columns

df = df.pct_change()
df = df.reset_index()
df = pd.melt(df, id_vars='Year',  var_name='county', value_name='Rate')
df
```

```python
#Steps to make data ready
# 1) Pull data and drop na if there are many nas, in this case I dropped more than 40
# 2) Transpose or Pivot to convert rows into columns. 
#this was needed because the columns had the years 2010, 2011, 2012, etc 
# 3) Other computations like playing with Date and converting data time
#Only when the index is date time, it is safe to apply time series functions
# 4) Melt the data again to get the final three columns of date, county and and the row you are looking for
# 5) Join the data using inner join on two columns
Zip_Zhvi_AllHomes_county = pd.read_csv('County_MedianListingPricePerSqft_AllHomes.csv')
Zip_Zhvi_AllHomes_county.dropna(thresh=40,inplace= True)  
Zip_Zhvi_AllHomes_county["RegionName"] = Zip_Zhvi_AllHomes_county["RegionName"] + ' '+ Zip_Zhvi_AllHomes_county["State"]
Zip_Zhvi_AllHomes_countyT= Zip_Zhvi_AllHomes_county.transpose()
Zip_Zhvi_AllHomes_countyT.drop(Zip_Zhvi_AllHomes_countyT.index[1:6], inplace= True)
Zip_Zhvi_AllHomes_countyT.columns = Zip_Zhvi_AllHomes_countyT.iloc[0]
Zip_Zhvi_AllHomes_countyT.drop(Zip_Zhvi_AllHomes_countyT.index[0], inplace= True)
#Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.reset_index()
Zip_Zhvi_AllHomes_countyT.index =  pd.to_datetime(Zip_Zhvi_AllHomes_countyT.index, format='%Y-%m')
Zip_Zhvi_AllHomes_countyT['month'] = pd.DatetimeIndex(Zip_Zhvi_AllHomes_countyT.index).month
# Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.loc[Zip_Zhvi_AllHomes_countyT['month'] == 1]
Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.pct_change()
del Zip_Zhvi_AllHomes_countyT['month']
Zip_Zhvi_AllHomes_countyT.head()
Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.reset_index()
Zip_Zhvi_AllHomes_countyT = pd.melt(Zip_Zhvi_AllHomes_countyT,  id_vars='index', var_name='county', value_name='price')
Zip_Zhvi_AllHomes_countyT = Zip_Zhvi_AllHomes_countyT.dropna()
#Zip_Zhvi_AllHomes_countyT.columns.values[0] = 'Year'
Zip_Zhvi_AllHomes_countyT['Year'] = Zip_Zhvi_AllHomes_countyT['index']
del Zip_Zhvi_AllHomes_countyT['index']
Zip_Zhvi_AllHomes_countyT['Year'].head()
```
